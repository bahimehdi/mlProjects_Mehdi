# üìã SESSION 3 : Building & Trusting Your Model (45 min)## Part 1: The Split (10 min)### üìò Theory: Train/Test Split**Probl√®me :** Si nous testons le mod√®le sur les m√™mes donn√©es qu'il a vues pendant l'entra√Ænement, c'est comme **donner les r√©ponses d'un examen √† un √©tudiant** !**Solution :** Diviser les donn√©es :- **80% pour l'entra√Ænement** (Train) : Le mod√®le apprend ici- **20% pour le test** (Test) : Le mod√®le n'a jamais vu ces donn√©es, test honn√™te !> **üí° Analogie :** Vous √©tudiez avec 80% des exercices d'un livre. L'examen final contient les 20% restants que vous n'avez jamais vus !---### üìò Theory: StratificationNotre dataset est **imbalanc√©** (plus de Broken que de Functional).**Stratification** = Garder les m√™mes proportions dans Train et Test.

```python
from sklearn.model_selection import train_test_split# Split avec stratificationX_train, X_test, y_train, y_test = train_test_split(    X, y,     test_size=0.2,      # 20% pour le test    random_state=42,     # Pour la reproductibilit√©    stratify=y           # Garder les proportions)print("‚úÖ Donn√©es divis√©es !")print(f"\nTrain set : {X_train.shape[0]} √©chantillons")print(f"Test set  : {X_test.shape[0]} √©chantillons")print(f"\nüìä Distribution dans Train : {y_train.value_counts().to_dict()}")print(f"üìä Distribution dans Test  : {y_test.value_counts().to_dict()}")
```

---## Part 2: Training (15 min)### üìò Theory: Random Forest Classifier**Random Forest** = Une "for√™t" de plusieurs arbres de d√©cision qui votent ensemble.**Pourquoi c'est puissant ?**- üå≥ Chaque arbre apprend diff√©remment- üó≥Ô∏è Le vote majoritaire r√©duit les erreurs- üí™ Fonctionne bien sur des donn√©es imbalanc√©es avec `class_weight='balanced'`---### üìò Theory: Class WeightNotre dataset a **60% Broken, 40% Functional**.Sans ajustement, le mod√®le va favoriser la classe majoritaire (Broken).**`class_weight='balanced'`** donne plus d'**importance** aux exemples de la classe minoritaire pendant l'entra√Ænement.

```python
from sklearn.ensemble import RandomForestClassifier# Cr√©er le mod√®le avec class_weight='balanced'model = RandomForestClassifier(    n_estimators=100,           # 100 arbres dans la for√™t    class_weight='balanced',    # Ajuster pour l'imbalance    random_state=42,    max_depth=10                # Limiter la profondeur pour √©viter l'overfitting)# Entra√Æner le mod√®leprint("üöÄ Entra√Ænement du mod√®le...")model.fit(X_train, y_train)print("‚úÖ Mod√®le entra√Æn√© avec succ√®s !")# Faire des pr√©dictionsy_pred = model.predict(X_test)print(f"\nüìä Pr√©dictions r√©alis√©es sur {len(y_pred)} pompes de test")
```

## Part 3: Evaluation (20 min)### üìò Theory: Les M√©triques pour la ClassificationPour un probl√®me de classification imbalanc√©e, **l'Accuracy n'est PAS suffisante** !**Les 4 m√©triques cl√©s :**1. **Precision (Pr√©cision)** : Parmi les pompes qu'on a pr√©dit "Broken", combien le sont vraiment ?   - Formule : TP / (TP + FP)   - Important si les **Faux Positifs** co√ªtent cher2. **Recall (Rappel)** : Parmi les pompes r√©ellement "Broken", combien avons-nous d√©tect√© ?   - Formule : TP / (TP + FN)   - Important si les **Faux N√©gatifs** co√ªtent cher (manquer une pompe cass√©e = pas d'eau !)3. **F1-Score** : Moyenne harmonique de Precision et Recall   - Formule : 2 √ó (Precision √ó Recall) / (Precision + Recall)   - **M√âTRIQUE PRIORITAIRE** pour ce projet4. **Confusion Matrix** : Tableau montrant TP, TN, FP, FN> **üí° Dans notre contexte :**> - **Faux N√©gatif** = Dire qu'une pompe cass√©e est fonctionnelle ‚Üí Les gens n'auront pas d'eau !> - **Faux Positif** = Dire qu'une pompe fonctionnelle est cass√©e ‚Üí Maintenance inutile (moins grave)

```python
from sklearn.metrics import f1_score, classification_report, confusion_matrix# 1. F1-Score (PRIORITAIRE)f1 = f1_score(y_test, y_pred)print(f"üéØ F1-Score : {f1:.3f}")print("(Plus c'est proche de 1.0, mieux c'est !)")# 2. Rapport de classification completprint("\nüìä Rapport de Classification :")print(classification_report(y_test, y_pred, target_names=['Functional', 'Broken']))
```
